## **MODELOS DE REGRESIÓN LINEAL DE ALTURA Y PESOS**

```python
# Importar librerias
import pandas as pd
import seaborn as sb
from sklearn.linear_model import LinearRegression
```

```python
# Cargar data
datos = pd.read_csv("C:\\Users\\ALEX\\OneDrive\\Escritorio\\Introducción a la IA\\peso_altura.csv")
datos
```

```python
# Graficar data
sb.scatterplot(x="Height", y="Weight", data=datos, hue="Weight", palette="coolwarm")
```

```python
# Separar datos en características (X), etiqueta (Y)
x = datos['Height']
y = datos['Weight']
```

```python
# Tranformar datos
x_procesada = x.values.reshape(-1,1)
y_procesada = y.values.reshape(-1,1)
```

```python
# Crear Modelo
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
```

```python
# Entrenar
modelo.fit(x_procesada, y_procesada)
```

```python
# Primera predicción altura de 1.73
modelo.predict([[1.73]])
```

```python
# Segunda predicción altura de 1.55
modelo.predict([[1.55]])
```

```python
# Evaluar al modelo
modelo.score(x_procesada,y_procesada)
```
